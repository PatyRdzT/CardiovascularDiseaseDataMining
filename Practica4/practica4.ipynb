# Práctica 4: Pruebas Estadísticas

Para esta cuarta práctica, se ejecuta código para realizar pruebas estadísticas en el dataset.

## Librerías

Se importan las librerías necesarias para las pruebas estadísticas.

```python
import pandas as pd
from scipy import stats
from scipy.stats import shapiro
```

## Importación de los datos desde CSV

Se crea una copia del dataset limpio para manejarlo y realizar cálculos en esa copia.

```python
# Se carga el conjunto de datos limpio.
df_original = pd.read_csv("C:\\Users\\Paty-\\OneDrive\\Documentos\\04 Facultad\\Séptimo Semestre\\Minería de Datos\\CardiovascularDiseaseDataMining\\Practica1\\cardiovascular_disease_cleaned.csv", sep=";")

# Se crea una copia del conjunto de datos limpio para trabajar en ella.
df = df_original.copy()

# Se muestran las primeras filas del conjunto de datos.
df.head()
```

## Formato de clases y variables a usar

Se separan en dos clases los valores de la variable cardio para tener un mejor manejo en grupos.

```python
cardio_yes = df[df['cardio'] == 'Yes']
cardio_no = df[df['cardio'] == 'No']
```

Se toman en cuenta las variables numéricas no escaladas para realizar las pruebas.

```python
numerical_vars = [
    'age', 'height', 'weight',
    'ap_hi', 'ap_lo'
]
```

## Verificación de la normalidad de los datos

Se verifica si los datos siguen o no una distribución normal para determinar si usar Kruskall Wallis.

```python
for var in numerical_vars:
    
    stat, p = shapiro(df[var])
    
    print(f"\nVariable: {var}")
    print("Statistic:", stat)
    print("p-value:", p)
    
    if p < 0.05:
        print("Resultado: No sigue una distribución normal")
    else:
        print("Resultado: Sigue una distribución normal")
```

## Prueba de Kruskal-Wallis

Dado que los datos no tienen una distribución normal, primero se hace la prueba Kruskal-Wallis.

```python
for var in ['age','height','weight','ap_hi','ap_lo']:
    
    stat, p = stats.kruskal(cardio_yes[var], cardio_no[var])

    print(f"\nVariable: {var}")
    print("Statistic:", stat)
    print("p-value:", p)

    if p < 0.05:
        print("Resultado: Diferencia significativa entre clases.")
    else:
        print("Resultado: No hay diferencia significativa entre clases.")
```

## T-Test

Se realiza la prueba T para averiguar si las medias entre clases son diferentes.

```python
for var in numerical_vars:
    
    stat, p = stats.ttest_ind(cardio_yes[var], cardio_no[var])
    
    print(f"\nVariable: {var}")
    print("T statistic:", stat)
    print("p-value:", p)
    
    if p < 0.05:
        print("Resultado: Diferencia significativa entre clases.")
    else:
        print("Resultado: No hay diferencia significativa entre clases.")
```

## ANOVA

Se aplica la prueba ANOVA para compararla con la T-Test.

```python
for var in numerical_vars:

    stat, p = stats.f_oneway(cardio_yes[var], cardio_no[var])

    print(f"\nVariable: {var}")
    print("F-statistic:", stat)
    print("p-value:", p)

    if p < 0.05:
        print("Resultado: Diferencia significativa entre clases.")
    else:
        print("Resultado: No hay diferencia significativa entre clases.")
```

## Conclusión

Los resultados de las tres pruebas realizadas indican que las variables numéricas analizadas presentan diferencias significativas entre las personas con y sin enfermedad cardiovascular, lo cual demuestra que las personas enfermas presentan diferencias medibles en variables relacionadas a su estado de salud.
